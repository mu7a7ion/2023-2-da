```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4], [10,20,30,40], label='ascending')
plt.plot([1,2,3,4], [40,50,20,10], label='descending')

plt.title("legend")
plt.xlabel('x.axis')
plt.ylabel('y.axis')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt


plt.plot([1,2,3,4], [10,20,30,40], color='r', linestyle='--', label='thick line', marker='d')
plt.plot([1,2,3,4], [40,50,20,10], label='descending')

plt.title('Linewidth')
plt.xlabel('x.axis')
plt.ylabel('y.axis')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)
plt.plot(x, y)
plt.grid(True)

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

plt.title('  ')
```

```python
import numpy as np
import matplotlib.pyplot as plt
N = 8
y = np.zeros(N)

x1 = np.linspace(0, 10, N, endpoint=True)
x2 = np.linspace(0, 10, N, endpoint=False)
plt.plot(x1, y, 'o')
plt.plot(x2, y + 0.5, 'o')
plt.ylim([-0.5, 1])

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
N = 8
y = np.zeros(N)

x1 = np.linspace(0, 10, N, endpoint=True)
x2 = np.linspace(0, 10, N, endpoint=False)
plt.plot(x1, y, 'o-')
plt.plot(x2, y + 0.5, 'o:')
plt.ylim([-0.5, 1])

plt.show()
```
